```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading a dataset:
phone_dataframe = pd.read_csv('C:/Users/HP USER/Downloads/archive/processed_data.csv')
```

```python
phone_dataframe
```

```python
plt.plot(phone_dataframe["phone_model"], phone_dataframe["storage"])
plt.title("Storage vs Phone Model")
plt.xlabel("Phone Model")
plt.ylabel("Storage (GB)")
plt.show()
```

```python
plt.grid(True)
plt.scatter(phone_dataframe["storage"], phone_dataframe["ram"])
plt.title("Phone Storage vs RAM by Brand")
plt.xlabel("Storage (GB)")
plt.ylabel("RAM (GB)")
plt.show()
```

```python
# Create a histogram for the 'BATTERY' column
plt.figure(figsize=(8, 6))
sns.histplot(data=phone_dataframe, x='BATTERY', kde=True, bins=20, color='blue')

# Set the title and labels
plt.title("Battery Capacity Distribution")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Count")

plt.show()
```

```python

# Create a KDE plot for 'BATTERY' by 'Year'
plt.figure(figsize=(8, 6))
sns.kdeplot(data=phone_dataframe, x="BATTERY", hue="Year", fill=True, common_norm=False, palette="crest", alpha=0.2)

# Add labels and title
plt.title("Kernel Density Plot of Battery Capacity by Year")
plt.xlabel("Battery Capacity (mAh)")
plt.ylabel("Density")

plt.show()
```
